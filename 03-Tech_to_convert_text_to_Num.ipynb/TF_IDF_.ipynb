
# TF-IDF Vectorizer
 Term Frequency-Inverse Document Frequency (TF-IDF) is a technique used to convert text documents into numerical vectors.
 It takes into account the importance of each word in the document and also the rarity of the word in the entire corpus.
 TF-IDF is an extension of the BoW technique and is used to generate numerical vectors for text documents.
 It is widely used in NLP and text analysis applications.


```python

def tfidf_vectorizer(data):
    """
    This function generates a TF-IDF vectorizer for a given text document.
    
    Parameters:
    data (str): The text document.
    
    Returns:
    A TF-IDF vectorizer.
    """
    vectorizer = TfidfVectorizer()
    tfidf_vectorizer = vectorizer.fit_transform([data])
    return tfidf_vectorizer
```

```python
data = """The quick brown fox jumps over the lazy dog.
    Never jump over the lazy dog until you know if it's friendly.
    The dog barks at the fox, and the fox runs away.A quick fox is a smart fox.",
    "Dogs are loyal animals."""
```

```python
import re
def clean_text(text):
    text=text.lower()
    text=re.sub(r'[^a-z\s]',"",text)
    text=re.sub(r'[0-9]',"",text)
    return text
clean_data=clean_text(data)
```

```python
import nltk
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("punkt_tab")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
clean_data = word_tokenize(clean_data, "english")
```

```python
stop_word=stopwords.words("english")
```

```python
clean_data=[word for word in clean_data if word not in stop_word]
```

```python
lamiza=WordNetLemmatizer()
```

```python
clean_data=[lamiza.lemmatize(w) for w in clean_data]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_idf=TfidfVectorizer()
```

```python
result=tf_idf.fit_transform(clean_data)
```
