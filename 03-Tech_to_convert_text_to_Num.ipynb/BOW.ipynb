
# What is Bag of Words (BoW)
 Bag of Words is a simple technique used for text representation in NLP. It involves counting the occurrence of each word in a document and storing it in a dictionary or a vector. The resulting vector is known as a bag of words.


Complete impletation of Bag_of_words

```python
data = """The quick brown fox jumps over the lazy dog.
    """
```

```python
import nltk
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("punkt_tab")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', "", text)
    text = re.sub(r'[0-9]', "", text)
    return text
```

```python
stop_word=stopwords.words("english")
```

```python
clean_data = clean_text(data)
clean_data = word_tokenize(clean_data)
clean_data = [word for word in clean_data if word not in stop_word]
```

```python
lamiza=WordNetLemmatizer()
```

```python
clean_data=[lamiza.lemmatize(w) for w in clean_data]
```

```python
# bag of word
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow=CountVectorizer()
```

```python
Bag_of_word=bow.fit_transform(clean_data)
```

```python
Bag_of_word.toarray()
```
