
### Major Techniques to Convert Text into Numeric Vector

1. **Bag-of-Words (BoW)**: This is a simple and widely used technique where a text document is represented as a bag or a set of its word frequencies. The document is then converted into a numerical vector by counting the frequency of each word in the document.

2. **Term Frequency-Inverse Document Frequency (TF-IDF)**: This is an extension of the BoW technique. It takes into account the importance of each word in the document and also the rarity of the word in the entire corpus.

3. **Word Embeddings**: This is a deep learning based technique where each word in the document is represented as a dense vector in a high dimensional space. The vectors are then combined to form a single numerical vector representing the entire document.

4. **Language Models**: This is a statistical technique which is used to predict the next word in a sequence of text given the context of the previous words. The numerical vector is then generated by taking into account the probability of each word given the context.

5. **One-Hot Encoding**: This is a simple technique where each unique word in the document is represented as a binary vector. The vectors are then combined to form a single numerical vector representing the entire document.

6. **Doc2Vec**: This is a deep learning based technique which is used to generate numerical vectors for documents. It is an extension of the Word2Vec technique and is used to generate vectors for documents rather than individual words.
