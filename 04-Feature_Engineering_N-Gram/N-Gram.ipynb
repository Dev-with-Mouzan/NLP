
# What is N-gram

An N-gram is a contiguous sequence of n items from a given sample of text or speech. N-grams are used to analyze text data and to identify patterns and relationships in the data. The size of the sequence is determined by the value of n. For example, if n is equal to 2, the sequence is a bigram and if n is equal to 3, the sequence is a trigram.

N-grams are commonly used in text analysis and natural language processing applications, such as language modeling, sentiment analysis, and machine translation. They are also used in speech recognition and speech synthesis applications.

For example, if we have a sentence "The quick brown fox jumped over the lazy dog", the N-grams for n equal to 2 are:

* The quick
* quick brown
* brown fox
* fox jumped
* jumped over
* over the
* the lazy
* lazy dog

And the N-grams for n equal to 3 are:

* The quick brown
* quick brown fox
* brown fox jumped
* fox jumped over
* jumped over the
* over the lazy
* the lazy dog



# Why we use N-Gram

We use N-grams because they are a simple and powerful way to analyze text data and to identify patterns and relationships in the data. N-grams allow us to capture the context and relationships between words in a sentence, which can be useful in a wide range of natural language processing applications. They are also simple to compute and require minimal computational resources, making them a useful tool for text analysis tasks.
