
# What is Text Preprocessing?

Text preprocessing is the process of transforming raw text data into a format that is more suitable for analysis. It typically involves a series of steps such as tokenization, removing stop words, stemming or lemmatizing, and removing punctuation. The goal of text preprocessing is to reduce the dimensionality of the data and to remove noise and irrelevant information. By doing so, it can improve the accuracy of machine learning models and other text analysis techniques.

# Steps to make text clean:

1.**Lower Case**: Before proceding the whole text is converted in to lowerCase                                           
2. **Tokenization**: Breaking down text into individual words or tokens.                                                      
3. **Stopword removal**: Removing common words like "the", "and", etc. that do not carry much meaning.                              
4. **Stemming or Lemmatization**: Reducing words to their base form so that words with the same meaning are treated  
as the same word.                                                                                                                                  
5. **Removing punctuation**: Removing punctuation marks from the text.                                                 
6. **Removing numbers**: Removing numbers from the text.                                                               
7. **Removing special characters**: Removing special characters from the text.                                                 
8. **Removing whitespace**: Removing whitespace from the text.                                                         
9. **Removing URLs**: Removing URLs from the text.                                                                           
10. **Removing HTML tags**: Removing HTML tags from the text.                                                                 
11. **Removing emails**: Removing emails from the text.
