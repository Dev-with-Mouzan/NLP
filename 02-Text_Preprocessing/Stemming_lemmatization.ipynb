
# What is Stemming
Stemming is the process of reducing words to their base form, known as a stem. The stem is the form of the word that is common to all its inflected variants. For example, the words "run", "running", and "runner" are all reduced to the same stem "run" when stemmed. Stemming is a natural language processing technique that is used to reduce words to their root form so that words with the same meaning can be treated as the same word. It is commonly used in text analysis and information retrieval applications.


```python
# Eample of stemming using NLTK
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["running", "happiness", "jumping"]
stemmed_words = [stemmer.stem(word) for word in words]
print("Original words:", words)
print("Stemmed words:", stemmed_words)
```


# What is Lemmatization
Lemmatization is the process of reducing words to their base or root form, known as a lemma. The lemma is the form of the word that is common to all its inflected variants. For example, the words "running", "happiness", and "jumping" are all reduced to the same lemma "run", "happiness", and "jump" when lemmatized. Lemmatization is a natural language processing technique that is used to reduce words to their root form so that words with the same meaning can be treated as the same word. It is commonly used in text analysis and information retrieval applications.


```python
# Eample of stemming using NLTK
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
words = ["studies", "happiness", "cars"]
stemmed_words = [lem.lemmatize(word) for word in words]
print("Original words:", words)
print("Stemmed words:", stemmed_words)
```
