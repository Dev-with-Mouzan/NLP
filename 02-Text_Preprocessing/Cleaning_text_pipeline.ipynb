# Text cleaning pipeline

```python
my_string = """!Natural Language Processing (NLP) is a subfield of linguistics, computer science,
and artificial intelligence. The history of NLP dates back to the 1950s,
when Alan Turing published his article "Computing Machinery and Intelligence."
Modern NLP leverages deep learning, transformers, and large language models (LLMs)
to achieve state-of-the-art performance on tasks like translation, summarization,
and question answering. OpenAI's GPT-4 and Google's Gemini are the latest milestones!."""
```

```python
# step 1: convert to lowercase
my_string=my_string.lower()
```

```python
my_string
```

```python
# step 2: remove punctuation
import re
re_pun=re.sub(r'[^a-z\s]','',my_string)
```

```python
re_pun
```

```python
# step 3: remove Number
re_num=re.sub(r'[0-9]','',re_pun)
```

```python
re_num
```

```python
# step 4 : remove URL
re_url=re.sub(r'http\S+|www\S+','',re_num)
```

```python
re_url
```

```python
# step 5: remove emojis
import emoji
def remove_emoji(text):
    return emoji.replace_emoji(text, replace='')
cleaned_string = remove_emoji(re_url)
cleaned_string
```

```python
# step 6: remove stop words
from nltk.corpus import stopwords
```

```python
remove_stopword=[word for word in cleaned_string.split() if word not in stopwords.words('english')]
```

```python
remove_stopword
```

```python
# step 7: tokenization
from nltk.tokenize import word_tokenize
token=word_tokenize(cleaned_string)
```
