# Simple NLP pipeline

```python
my_string = "HELLO! This is a string containing punctuation, UPPERCASE words, and several interesting words ending in."
```

```python
# step 1: convert to lowercase
my_string=my_string.lower()
```

```python
my_string
```

```python
import re
```

```python
# step 2: remove punctuation
re_pun=re.sub(r'[^a-z\s]','',my_string)
```

```python
re_pun
```

```python
# step 3: remove Number
re_num=re.sub(r'[0-9]','',re_pun)
```

```python
re_num
```

```python
# step 4 : remove URL
re_url=re.sub(r'http\S+|www\S+','',re_num)
```
